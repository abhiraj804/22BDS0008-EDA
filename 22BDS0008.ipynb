```python
# K-means Clustering
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Select features for clustering
AbhirajDas_cluster_features = AbhirajDas_credit_card_data[['income', 'expenditure']]

# Scale the features
AbhirajDas_scaler = StandardScaler()
AbhirajDas_scaled_features = AbhirajDas_scaler.fit_transform(AbhirajDas_cluster_features)


# Apply K-Means with the optimal k
AbhirajDas_optimal_k = 4
AbhirajDas_kmeans = KMeans(n_clusters=AbhirajDas_optimal_k, random_state=42, n_init=10)
AbhirajDas_cluster_labels = AbhirajDas_kmeans.fit_predict(AbhirajDas_scaled_features)

# Add cluster labels to the original dataframe
AbhirajDas_credit_card_data['cluster'] = AbhirajDas_cluster_labels

# Visualize the clusters
plt.figure(figsize=(10, 7))
sns.scatterplot(data=AbhirajDas_credit_card_data, x='income', y='expenditure', hue='cluster', palette='viridis', s=50, legend='full')
plt.scatter(AbhirajDas_kmeans.cluster_centers_[:, 0], AbhirajDas_kmeans.cluster_centers_[:, 1], s=200, c='red', marker='X', label='Centroids')
plt.title(f'{AbhirajDas_roll_no} - Customer Segments (K-Means Clustering)')
plt.xlabel('Income')
plt.ylabel('Expenditure')
plt.legend()
plt.show()

# Display the first few rows with cluster assignments
print("\nData with Cluster Assignments:")
print(AbhirajDas_credit_card_data[['income', 'expenditure', 'cluster']].head())
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Selecting only numerical features for PCA
AbhirajDas_numerical_features = ['reports', 'age', 'income', 'share', 'expenditure', 'dependents', 'months', 'active']
AbhirajDas_numerical_data = AbhirajDas_credit_card_data[AbhirajDas_numerical_features]

# Standardizing the features
AbhirajDas_scaler = StandardScaler()
AbhirajDas_scaled_data = AbhirajDas_scaler.fit_transform(AbhirajDas_numerical_data)

# Applying PCA
AbhirajDas_pca = PCA(n_components=2)  # Reducing to 2 components for visualization
AbhirajDas_principal_components = AbhirajDas_pca.fit_transform(AbhirajDas_scaled_data)

# Creating a DataFrame with the principal components
AbhirajDas_pca_df = pd.DataFrame(data=AbhirajDas_principal_components, columns=['PC1', 'PC2'])

print("First 5 rows of the PCA components:")
print(AbhirajDas_pca_df.head())

# Explained variance
print(f"\nExplained variance by PC1: {AbhirajDas_pca.explained_variance_ratio_[0]:.4f}")
print(f"Explained variance by PC2: {AbhirajDas_pca.explained_variance_ratio_[1]:.4f}")
print(f"Total explained variance: {AbhirajDas_pca.explained_variance_ratio_.sum():.4f}")

# Plotting the PCA components
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PC1', y='PC2', data=AbhirajDas_pca_df)
plt.title(f'{AbhirajDas_roll_no} - PCA: PC1 vs PC2')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
```

```python
#Model Building
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Preprocessing: Encoding categorical variables
AbhirajDas_model_data = AbhirajDas_credit_card_data.drop(columns=['rownames'])
AbhirajDas_categorical_features = ['card', 'owner', 'selfemp']

AbhirajDas_label_encoders = {}
for col in AbhirajDas_categorical_features:
    le = LabelEncoder()
    AbhirajDas_model_data[col] = le.fit_transform(AbhirajDas_model_data[col])
    AbhirajDas_label_encoders[col] = le

# Defining features (X) and target (y)
AbhirajDas_X = AbhirajDas_model_data.drop(columns=['card'])
AbhirajDas_y = AbhirajDas_model_data['card']

# Splitting the data
AbhirajDas_X_train, AbhirajDas_X_test, AbhirajDas_y_train, AbhirajDas_y_test = train_test_split(
    AbhirajDas_X, AbhirajDas_y, test_size=0.3, random_state=42
)

# Standardizing numerical features
AbhirajDas_features_to_scale = [col for col in AbhirajDas_numerical_features if col in AbhirajDas_X.columns]

scaler = StandardScaler()
AbhirajDas_X_train[AbhirajDas_features_to_scale] = scaler.fit_transform(AbhirajDas_X_train[AbhirajDas_features_to_scale])
AbhirajDas_X_test[AbhirajDas_features_to_scale] = scaler.transform(AbhirajDas_X_test[AbhirajDas_features_to_scale])

# Building the model (Logistic Regression for classification)
AbhirajDas_model = LogisticRegression(random_state=42, max_iter=1000)
AbhirajDas_model.fit(AbhirajDas_X_train, AbhirajDas_y_train)

# Making predictions
AbhirajDas_y_pred = AbhirajDas_model.predict(AbhirajDas_X_test)

# Evaluating the model
AbhirajDas_accuracy = accuracy_score(AbhirajDas_y_test, AbhirajDas_y_pred)
AbhirajDas_conf_matrix = confusion_matrix(AbhirajDas_y_test, AbhirajDas_y_pred)
AbhirajDas_class_report = classification_report(AbhirajDas_y_test, AbhirajDas_y_pred)

print(f"Logistic Regression (Target: 'card')\n")
print(f"Accuracy: {AbhirajDas_accuracy:.4f}\n")
print("Confusion Matrix:")
print(AbhirajDas_conf_matrix)
print("\nClassification Report:")
print(AbhirajDas_class_report)
```
