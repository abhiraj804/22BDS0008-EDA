# **Data Cleaning**

```python
#9.1)
# Checking for missing values to get non-numeric values.
AbhirajDas_age_non_numeric_count = AbhirajDas_credit_card_data['age'].isnull().sum()
AbhirajDas_income_non_numeric_count = AbhirajDas_credit_card_data['income'].isnull().sum()
AbhirajDas_expenditure_non_numeric_count = AbhirajDas_credit_card_data['expenditure'].isnull().sum()

#count of missing values
print("Number of non-numeric/missing values found:")
print(f"age: {AbhirajDas_age_non_numeric_count}")
print(f"income: {AbhirajDas_income_non_numeric_count}")
print(f"expenditure: {AbhirajDas_expenditure_non_numeric_count}")
```

```python
#9.2)
#As there are no missing values to do this question wel will create some missing values and demonstrate the question.

# Create a copy of the 'expenditure' column
AbhirajDas_expenditure_demo = AbhirajDas_credit_card_data['expenditure'].copy()

# Adding some missing values (NaNs).
AbhirajDas_expenditure_demo.iloc[[2, 4, 8]] = np.nan
print("\n'expenditure' column with artificial missing values added:")
print(AbhirajDas_expenditure_demo.head(10))

# Mean of the expenditure.
AbhirajDas_expenditure_mean = AbhirajDas_expenditure_demo.mean()
print(f"\nCalculated Mean of 'expenditure': {AbhirajDas_expenditure_mean:.2f}")

# Replacing the missing values
AbhirajDas_expenditure_demo.fillna(AbhirajDas_expenditure_mean, inplace=True)

# Converting the column to a numeric data type
AbhirajDas_expenditure_demo = pd.to_numeric(AbhirajDas_expenditure_demo)

# Printing the column after filling the missing values
print("\n'expenditure' column after replacing missing values with the mean:")
print(AbhirajDas_expenditure_demo.head(10))

# Confirming no missing values left
print(f"\nNumber of missing values remaining: {AbhirajDas_expenditure_demo.isnull().sum()}")
```

# **Statistical Computation**

```python
#1. Measures of Central Tendency for income
AbhirajDas_income_mean = AbhirajDas_credit_card_data['income'].mean()
AbhirajDas_income_median = AbhirajDas_credit_card_data['income'].median()
AbhirajDas_income_mode = AbhirajDas_credit_card_data['income'].mode()[0]

print("Measures of Central Tendency for income")
print(f"Mean: {AbhirajDas_income_mean:.2f}")
print(f"Median: {AbhirajDas_income_median:.2f}")
print(f"Mode: {AbhirajDas_income_mode:.2f}")
```

```python
#Measures of Dispersion for income
AbhirajDas_income_variance = AbhirajDas_credit_card_data['income'].var()
AbhirajDas_income_std = AbhirajDas_credit_card_data['income'].std()
AbhirajDas_income_range = AbhirajDas_credit_card_data['income'].max() - AbhirajDas_credit_card_data['income'].min()

print("Measures of Dispersion for income")
print(f"Variance: {AbhirajDas_income_variance:.2f}")
print(f"Standard Deviation: {AbhirajDas_income_std:.2f}")
print(f"Range: {AbhirajDas_income_range:.2f}")
```

```python
#3. Quartiles and Interquartile Range (IQR) for income
AbhirajDas_income_q1 = AbhirajDas_credit_card_data['income'].quantile(0.25)
AbhirajDas_income_q3 = AbhirajDas_credit_card_data['income'].quantile(0.75)
AbhirajDas_income_iqr = AbhirajDas_income_q3 - AbhirajDas_income_q1

print("Quartiles and IQR for income")
print(f"First Quartile (Q1): {AbhirajDas_income_q1:.2f}")
print(f"Third Quartile (Q3): {AbhirajDas_income_q3:.2f}")
print(f"Interquartile Range (IQR): {AbhirajDas_income_iqr:.2f}")
```

```python
#4. Correlation between income and expenditure
AbhirajDas_correlation = AbhirajDas_credit_card_data['income'].corr(AbhirajDas_credit_card_data['expenditure'])

print("Correlation")
print(f"Correlation between 'income' and 'expenditure': {AbhirajDas_correlation:.4f}")
```
